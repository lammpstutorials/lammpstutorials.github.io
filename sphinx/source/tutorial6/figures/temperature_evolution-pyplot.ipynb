```python
import numpy as np
import sys, os, git, lammps_logfile
import warnings
warnings.filterwarnings("ignore")
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
path_in_folder = current_path[len(git_path)+1:]
level = path_in_folder.split("/")[-2]
tutorial_name = path_in_folder.split("/")[-1]
print("level:" , level, "& tutorial name:", tutorial_name)
sys.path.append(git_path + "/docs/sphinx/source/tutorials/figures/pyplot-perso")
from plttools import PltTools
path_figures = current_path[len(git_path):] + '/'
data_path = git_path + "/docs/lammpstutorials-inputs/" + level + "/" + tutorial_name + "/SilicaBlock/"
print("data path: ", data_path)
```

```python
temperature = np.loadtxt(data_path + "temperature.dat")
time, temperature = temperature.T
time /= 1000 # ps
```

```python
filename = "temperature_evolution"
for dark_mode in [False, True]:
    myplt = PltTools()
    myplt.prepare_figure(fig_size = (18,6), dark_mode = dark_mode,
                        transparency = True, use_serif=True, n_line=1)
    myplt.add_panel()
    myplt.add_plot(x = time, y = temperature, linewidth_data = 3,
                   marker = "-", data_color = 0, markersize = 12)
    x = np.arange(-100, 8100)
    myplt.add_plot(x = x*0+50, y = x, marker= '--', data_color="autogray", linewidth=3.5)
    myplt.add_plot(x = x*0+100, y = x, marker= '--', data_color="autogray", linewidth=3.5)
    myplt.add_plot(x = x*0+300, y = x, marker= '--', data_color="autogray", linewidth=3.5)
    myplt.complete_panel(ylabel = r'$T ~ \mathrm{(K)}$',
                         xlabel = r'$t ~ \mathrm{(ps)}$',
                         xpad = 10, legend=False, handlelength_legend=1)
    #myplt.set_boundaries(x_ticks=np.arange(-2., 2.2, 0.5), x_boundaries=(-2.3, 2.3),
    #               y_ticks=np.arange(0, 0.033, 0.01))
    #myplt.add_subplotlabels()
    myplt.save_figure(filename = filename, saving_path = './')
```

```python
dimensions = np.loadtxt(data_path + "dimensions.dat")
time, volume, lx, ly, lz = dimensions.T
time /= 1000 # ps
```

```python
filename = "dimensions_evolution"
for dark_mode in [False, True]:
    myplt = PltTools()
    myplt.prepare_figure(fig_size = (18,6), dark_mode = dark_mode,
                        transparency = True, use_serif=True, n_line=1)
    myplt.add_panel()
    myplt.add_plot(x = time, y = lx, linewidth_data = 3,
                   marker = "-", data_color = 0, markersize = 12, data_label = r'$\ell_x$')
    myplt.add_plot(x = time, y = ly, linewidth_data = 3,
                   marker = "-", data_color = 1, markersize = 12, data_label = r'$\ell_y$')
    myplt.add_plot(x = time, y = lz, linewidth_data = 3,
                   marker = "-", data_color = "autogray", markersize = 12, data_label = r'$\ell_z$')
    x = np.arange(8, 35)
    myplt.add_plot(x = x*0+50, y = x, marker= '--', data_color="autogray", linewidth=3.5)
    myplt.add_plot(x = x*0+100, y = x, marker= '--', data_color="autogray", linewidth=3.5)
    myplt.add_plot(x = x*0+300, y = x, marker= '--', data_color="autogray", linewidth=3.5)
    myplt.complete_panel(ylabel = r'$\mathrm{box ~ size} ~ (\mathrm{\AA})$',
                         xlabel = r'$t ~ \mathrm{(ps)}$',
                         xpad = 10, legend=True, handlelength_legend=1)
    myplt.set_boundaries(y_ticks=np.arange(10, 35, 5), y_boundaries=(8, 32))
    #myplt.add_subplotlabels()
    myplt.save_figure(filename = filename, saving_path = './')
```
