```python
import numpy as np
import sys, os, git
from matplotlib import pyplot as plt
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
path_in_folder = current_path[len(git_path)+1:]
level = path_in_folder.split("/")[-2]
tutorial_name = path_in_folder.split("/")[-1]
print("level:" , level, "& tutorial name:", tutorial_name)
sys.path.append(git_path + "/docs/sphinx/source/tutorials/figures/pyplot-perso/")
from functions import complete_panel, save_figure, set_boundaries, \
                      add_subplotlabels, set_boundaries, prepare_figure
from color_series1 import colors
path_figures = current_path[len(git_path):] + '/'
data_path = git_path + "/docs/lammpstutorials-inputs/" + level + "/" + tutorial_name + "/SilicaBlock/"
print("data path: ", data_path)
```

```python
temperature = np.loadtxt(data_path + "temperature.dat")
time, temperature = temperature.T
time /= 1000 # ps
```

```python
filename = "temperature_evolution"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,6))
    ax, n, l_tot, c_tot = [], 0, 1, 1
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    ax[-1].plot(time, temperature, color=colors["myblue"], linewidth=3)
    line = np.arange(-100, 8100)
    ax[-1].plot(line*0 + 50, line, '--', linewidth = 2, color=mygray)
    ax[-1].plot(line*0 + 100, line, '--', linewidth = 2, color=mygray)
    ax[-1].plot(line*0 + 300, line, '--', linewidth = 2, color=mygray)
    complete_panel(ax[-1], r'$t ~ \textrm{(ps)}$', r'$T ~ \textrm{(K)}$',
                legend=False, axis_color=mygray, xpad=8)
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```

```python
dimensions = np.loadtxt(data_path + "dimensions.dat")
time, volume, lx, ly, lz = dimensions.T
time /= 1000 # ps

filename = "dimensions_evolution"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):    
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,6))
    ax, n, l_tot, c_tot = [], 0, 1, 1
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    ax[-1].plot(time, lx, color=colors["myblue"], linewidth=3, label=r'$\ell_x$')
    ax[-1].plot(time, ly, color=colors["myorange"], linewidth=3, label=r'$\ell_y$')
    ax[-1].plot(time, lz, color=mygray, linewidth=3, label=r'$\ell_z$')
    line = np.arange(10, 30)
    ax[-1].plot(line*0 + 50, line, '--', linewidth = 2, color=mygray)
    ax[-1].plot(line*0 + 100, line, '--', linewidth = 2, color=mygray)
    ax[-1].plot(line*0 + 300, line, '--', linewidth = 2, color=mygray)
    complete_panel(ax[-1], r'$t ~ \textrm{(ps)}$', r'$\textrm{box ~ size} ~ (\textrm{\AA})$',
            legend=True, axis_color=mygray, xpad=8)
    set_boundaries(plt, y_ticks=np.arange(10, 35, 5))
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```

```python
filename = "temperature-article"
mode = 'light'
mygray = colors["mylightgray"]
fig = prepare_figure(mode, transparency = False, desired_figsize=(12,5), serif=False)
ax, n, l_tot, c_tot = [], 0, 1, 1
n += 1
ax.append(plt.subplot(l_tot, c_tot, n))
ax[-1].plot(time[::6], temperature[::6], 'o', color=colors["unisoncyan"], linewidth=3, markersize=12)
line = np.arange(-100, 8100)
ax[-1].plot(line*0 + 50, line, '--', linewidth = 2, color=mygray)
ax[-1].plot(line*0 + 100, line, '--', linewidth = 2, color=mygray)
ax[-1].plot(line*0 + 300, line, '--', linewidth = 2, color=mygray)
complete_panel(ax[-1], r'$t ~ \mathrm{(ps)}$', r'$T ~ \mathrm{(K)}$',
            legend=False, axis_color=mygray, xpad=8)
save_figure(plt, fig, mode, git_path, path_figures, filename)
```
