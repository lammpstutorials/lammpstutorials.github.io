```python
import numpy as np
import sys, os, git
from matplotlib import pyplot as plt
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
path_in_folder = current_path[len(git_path)+1:]
level = path_in_folder.split("/")[-2]
tutorial_name = path_in_folder.split("/")[-1]
print("level:" , level, "& tutorial name:", tutorial_name)
sys.path.append(git_path + "/docs/sphinx/source/tutorials/figures/pyplot-perso/")
from functions import complete_panel, save_figure, set_boundaries, \
                      add_subplotlabels, set_boundaries, prepare_figure
from color_series1 import colors
path_figures = current_path[len(git_path):] + '/'
data_path = git_path + "/docs/lammpstutorials-inputs/" + level + "/" + tutorial_name + "/unbreakable-bonds/"
print("data path: ", data_path)
```

```python
length = np.loadtxt(data_path + "cnt_length.dat")
time, length = length.T
time /= 1000 # ps
time_0 = time[time<=5]
time_1 = time[time>5]
length_0 = length[time<=5]
length_1 = length[time>5]
```

```python
filename = "length-unbreakable"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):    
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,5))
    ax, n, l_tot, c_tot = [], 0, 1, 1
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    #line = np.arange(50, 70, 0.1)
    #ax[-1].plot(line*0+5, line, '--', color=mygray, linewidth=3.5)
    ax[-1].plot(time_0, length_0, color=colors["myorange"], linewidth=3.5)
    ax[-1].plot(time_1, length_1, color=colors["myblue"], linewidth=3.5)
    complete_panel(ax[-1], r'$t ~ \mathrm{(ps)}$', r'$L \mathrm{(\AA)}$',
                    legend=False, axis_color=mygray, locator_y=3, xpad=8)
    set_boundaries(plt, x_ticks=np.arange(0, 18, 2), y_ticks=np.arange(55, 68, 3),
                   x_boundaries=(-0.8, 16.2), y_boundaries=(54, 68))
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```
